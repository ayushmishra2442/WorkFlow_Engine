# [auth].[sp_GetRolesForUser]

```sql
/****** Object:  StoredProcedure [auth].[sp_GetRolesForUser]    Script Date: 6/14/2026 6:23:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================================
-- SP: auth.sp_GetRolesForUser
-- All active role assignments for a given user
-- JOINs auth.Roles to return RoleName
-- ============================================================
CREATE   PROCEDURE [auth].[sp_GetRolesForUser]
    @UserId UNIQUEIDENTIFIER
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        ur.UserRoleId,
        ur.UserId,
        ur.RoleId,
        r.Name  AS RoleName,
        ur.AssignedOn
    FROM auth.UserRoles ur
    INNER JOIN auth.Roles r
        ON r.RoleId     = ur.RoleId
        AND r.DeleteFlag = 0
    WHERE ur.UserId     = @UserId
      AND ur.DeleteFlag  = 0
    ORDER BY r.Name ASC;
END;

GO
```

# [auth].[sp_GetUsers]

```sql
/****** Object:  StoredProcedure [auth].[sp_GetUsers]    Script Date: 6/14/2026 6:23:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Update sp_GetUsers to include TotalCount
CREATE   PROCEDURE [auth].[sp_GetUsers]
    @PageNumber INT = 1,
    @PageSize   INT = 10
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        COUNT(*) OVER()     AS TotalCount,
        UserId,
        OrganizationId,
        DisplayName,
        Email,
        IsActive,
        CreatedOn
    FROM auth.Users
    WHERE DeleteFlag = 0
    ORDER BY DisplayName ASC
    OFFSET (@PageNumber - 1) * @PageSize ROWS
    FETCH NEXT @PageSize ROWS ONLY;
END;

GO
```
